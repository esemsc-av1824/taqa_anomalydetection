```python
import pandas as pd
```

```python
p8_1 = pd.read_csv('/content/drive/MyDrive/TAQA/P8-1_streamSummary_2025-06-30.csv')
p8_7 = pd.read_csv('/content/drive/MyDrive/TAQA/P8-7_streamSummary_2025-07-01.csv')
p8_38 = pd.read_csv('/content/drive/MyDrive/TAQA/P8-38_streamSummary_2025-06-30.csv')
p8_41 = pd.read_csv('/content/drive/MyDrive/TAQA/P8-41_streamSummary_2025-06-30.csv')
p8_59 = pd.read_csv('/content/drive/MyDrive/TAQA/P8-59_streamSummary_2025-06-30.csv')
p8_1.head()
```

```python
frames = [p8_7, p8_1, p8_38,p8_41, p8_59]
p8_full = pd.concat(frames, ignore_index=True)
p8_full['param'] = p8_full['Stream Id'].str.split('.').str[-1]
p8_full['tool'] = p8_full['Stream Id'].str.split('.').str[1]
#diplay p8 dims
p8_full
```

```python
p8_full.dtypes
```

```python
p8_full['% missing'] = p8_full['% missing'].str.replace('%','', regex=False).astype(float) #converting %
p8_full['Std'] = pd.to_numeric(p8_full['Std'], errors='coerce')
```

```python
filter1_p8 = p8_full[(p8_full['Std'] > 0) & (p8_full['% missing'] < 15)]
```

```python
filter1_p8
```

```python
tool_count   = filter1_p8.groupby('param')['tool'].nunique()
survivors    = tool_count[tool_count >= 5]

print(survivors)
```
